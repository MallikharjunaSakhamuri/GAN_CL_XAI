```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, roc_auc_score, confusion_matrix, classification_report
from sklearn.base import clone
from sklearn.inspection import permutation_importance
from sklearn.model_selection import train_test_split
import lime
import lime.lime_tabular
import warnings
from sklearn.preprocessing import StandardScaler
from rdkit import Chem
from rdkit.Chem import Descriptors, Lipinski, QED
import time

# Start timing the script
start_time = time.time()

# Suppress warnings
warnings.filterwarnings('ignore')

# Set up visualization parameters
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 8)

# Helper function for feature importance dataframe
def imp_df(column_names, importances):
    """Create a dataframe of feature importances"""
    df = pd.DataFrame({
        'feature': column_names,
        'feature_importance': importances
    }).sort_values('feature_importance', ascending=False).reset_index(drop=True)
    return df

# Visualization function
def var_imp_plot(imp_df, title, figsize=(12, 8)):
    """Plot feature importances as a horizontal bar chart"""
    plt.figure(figsize=figsize)
    ax = sns.barplot(x='feature_importance', y='feature', data=imp_df, orient='h', color='royalblue')
    
    # Add value labels to the bars
    for i, v in enumerate(imp_df['feature_importance']):
        ax.text(max(v + 0.001, 0.001), i, f'{v:.4f}', va='center')
    
    plt.title(title, fontsize=16)
    plt.tight_layout()
    plt.show()

# Function to analyze random feature position
def analyze_random_feature(imp_df):
    """Analyze where the random feature ranks in importance"""
    random_row = imp_df[imp_df['feature'] == 'random']
    if len(random_row) == 0:
        return "Random feature not found in results"
    
    random_rank = random_row.index[0] + 1
    random_importance = random_row['feature_importance'].values[0]
    total_features = len(imp_df)
    
    return {
        "rank": random_rank,
        "total_features": total_features, 
        "percentile": 100 * (total_features - random_rank) / total_features,
        "importance_value": random_importance
    }

# Function for LIME visualization
def plot_lime_explanation(exp_list, title):
    """Create a horizontal bar chart from LIME explanation list"""
    features = [item[0] for item in exp_list]
    importances = [item[1] for item in exp_list]
    
    df = pd.DataFrame({
        'feature': features,
        'importance': importances
    })
    df = df.sort_values('importance', key=abs, ascending=False)
    
    plt.figure(figsize=(12, 8))
    colors = ['royalblue' if x > 0 else 'lightcoral' for x in df['importance']]
    ax = sns.barplot(x='importance', y='feature', data=df, palette=colors)
    
    for i, v in enumerate(df['importance']):
        ax.text(v + 0.01 if v >= 0 else v - 0.11, i, f'{v:.4f}', va='center')
    
    plt.axvline(x=0, color='black', linestyle='-', linewidth=0.5)
    plt.title(title, fontsize=16)
    plt.tight_layout()
    plt.show()

# Drop column feature importance function for classification
def drop_col_feat_imp(model, X_train, y_train, metric='roc_auc', random_state=42):
    """Calculate feature importance by dropping columns one by one"""
    model_clone = clone(model)
    model_clone.random_state = random_state
    
    model_clone.fit(X_train, y_train)
    
    # Predict with full model for benchmark score
    y_pred_proba = model_clone.predict_proba(X_train)[:, 1]
    if metric == 'roc_auc':
        benchmark_score = roc_auc_score(y_train, y_pred_proba)
    elif metric == 'accuracy':
        y_pred = model_clone.predict(X_train)
        benchmark_score = accuracy_score(y_train, y_pred)
    else:
        raise ValueError("Metric must be 'roc_auc' or 'accuracy'")
    
    importances = []

    # Drop one column at a time and measure performance decrease
    for col in X_train.columns:
        model_clone = clone(model)
        model_clone.random_state = random_state
        X_drop = X_train.drop(col, axis=1)
        model_clone.fit(X_drop, y_train)
        
        # Calculate score with dropped column
        if metric == 'roc_auc':
            y_drop_pred_proba = model_clone.predict_proba(X_drop)[:, 1]
            drop_col_score = roc_auc_score(y_train, y_drop_pred_proba)
        else:  # accuracy
            y_drop_pred = model_clone.predict(X_drop)
            drop_col_score = accuracy_score(y_train, y_drop_pred)
            
        importances.append(benchmark_score - drop_col_score)

    return importances

# Function to calculate molecular descriptors from SMILES - Using only reliable RDKit functions
def calculate_mol_descriptors(smiles_list):
    """Calculate RDKit molecular descriptors from SMILES strings"""
    descriptors_list = []
    valid_indices = []
    
    for i, smiles in enumerate(smiles_list):
        try:
            # Sanitize the molecules with more lenient parameters
            mol = Chem.MolFromSmiles(smiles, sanitize=False)
            if mol is None:
                continue
                
            # Try to sanitize with error handling
            try:
                Chem.SanitizeMol(mol)
            except:
                # If full sanitization fails, try partial sanitization
                Chem.SanitizeMol(mol, sanitizeOps=Chem.SanitizeFlags.SANITIZE_FINDRADICALS|
                                          Chem.SanitizeFlags.SANITIZE_KEKULIZE|
                                          Chem.SanitizeFlags.SANITIZE_SETAROMATICITY|
                                          Chem.SanitizeFlags.SANITIZE_SETCONJUGATION|
                                          Chem.SanitizeFlags.SANITIZE_SETHYBRIDIZATION|
                                          Chem.SanitizeFlags.SANITIZE_SYMMRINGS,
                                 catchErrors=True)
            
            # Calculate basic physicochemical properties
            try:
                mw = Descriptors.MolWt(mol)
                logp = Descriptors.MolLogP(mol)
                hbd = Lipinski.NumHDonors(mol)
                hba = Lipinski.NumHAcceptors(mol)
                rotatable_bonds = Descriptors.NumRotatableBonds(mol)
                tpsa = Descriptors.TPSA(mol)
                num_rings = Descriptors.RingCount(mol)
                heavy_atoms = mol.GetNumHeavyAtoms()
                
                # Additional descriptors
                num_aromatic_rings = Lipinski.NumAromaticRings(mol)
                num_aliphatic_rings = Lipinski.NumAliphaticRings(mol)
                fraction_sp3 = Descriptors.FractionCSP3(mol)
                
                # Try to calculate QED - drug-likeness score
                try:
                    qed_value = QED.qed(mol)
                except:
                    qed_value = np.nan
                
                # Rule of 5 violations (Lipinski rules)
                ro5_violations = 0
                if mw > 500: ro5_violations += 1
                if logp > 5: ro5_violations += 1
                if hbd > 5: ro5_violations += 1
                if hba > 10: ro5_violations += 1
                
                # Additional calculated properties
                molar_refractivity = Descriptors.MolMR(mol)
                chi0v = Descriptors.Chi0v(mol)
                chi1v = Descriptors.Chi1v(mol)
                chi2v = Descriptors.Chi2v(mol)
                chi3v = Descriptors.Chi3v(mol)
                chi4v = Descriptors.Chi4v(mol)
                kappa1 = Descriptors.Kappa1(mol)
                kappa2 = Descriptors.Kappa2(mol)
                kappa3 = Descriptors.Kappa3(mol)
                
                # Drug-like properties
                max_partial_charge = Descriptors.MaxPartialCharge(mol) if hasattr(Descriptors, 'MaxPartialCharge') else 0
                min_partial_charge = Descriptors.MinPartialCharge(mol) if hasattr(Descriptors, 'MinPartialCharge') else 0
                num_heteroatoms = Lipinski.NumHeteroatoms(mol)
                num_stereocenters = Chem.rdMolDescriptors.CalcNumAtomStereoCenters(mol) if hasattr(Chem.rdMolDescriptors, 'CalcNumAtomStereoCenters') else 0
                
                # Toxicity-relevant properties (extended)
                num_nitrogen_atoms = sum(1 for atom in mol.GetAtoms() if atom.GetAtomicNum() == 7)
                num_oxygen_atoms = sum(1 for atom in mol.GetAtoms() if atom.GetAtomicNum() == 8)
                num_sulfur_atoms = sum(1 for atom in mol.GetAtoms() if atom.GetAtomicNum() == 16)
                num_halogen_atoms = sum(1 for atom in mol.GetAtoms() if atom.GetAtomicNum() in [9, 17, 35, 53])
                num_phosphorus_atoms = sum(1 for atom in mol.GetAtoms() if atom.GetAtomicNum() == 15)
                
                descriptors = {
                    # Basic properties
                    'MW': mw,
                    'LogP': logp,
                    'HBD': hbd,
                    'HBA': hba,
                    'RotatableBonds': rotatable_bonds,
                    'TPSA': tpsa,
                    'QED': qed_value,
                    'NumRings': num_rings,
                    'HeavyAtoms': heavy_atoms,
                    'RO5Violations': ro5_violations,
                    
                    # Additional properties
                    'AromaticRings': num_aromatic_rings,
                    'AliphaticRings': num_aliphatic_rings,
                    'FractionSP3': fraction_sp3,
                    'MolarRefractivity': molar_refractivity,
                    
                    # Connectivity indices
                    'Chi0v': chi0v,
                    'Chi1v': chi1v,
                    'Chi2v': chi2v,
                    'Chi3v': chi3v,
                    'Chi4v': chi4v,
                    
                    # Kappa shape indices
                    'Kappa1': kappa1,
                    'Kappa2': kappa2,
                    'Kappa3': kappa3,
                    
                    # Drug-like & Toxicity relevant properties
                    'MaxPartialCharge': max_partial_charge,
                    'MinPartialCharge': min_partial_charge,
                    'NumHeteroatoms': num_heteroatoms,
                    'NumStereocenters': num_stereocenters,
                    'NumNitrogenAtoms': num_nitrogen_atoms,
                    'NumOxygenAtoms': num_oxygen_atoms,
                    'NumSulfurAtoms': num_sulfur_atoms,
                    'NumHalogenAtoms': num_halogen_atoms,
                    'NumPhosphorusAtoms': num_phosphorus_atoms
                }
                
                descriptors_list.append(descriptors)
                valid_indices.append(i)
                
            except Exception as e:
                print(f"Error calculating basic descriptors for {smiles}: {e}")
                
        except Exception as e:
            print(f"Error parsing SMILES for {smiles}: {e}")
            
    return pd.DataFrame(descriptors_list), valid_indices

# Function to run the entire feature importance analysis for a given task
def analyze_feature_importance(X, y, target_name, random_state=42):
    """Run complete feature importance analysis for a classification task"""
    print(f"\n{'='*30}")
    print(f"ANALYZING {target_name.upper()}")
    print(f"{'='*30}")
    
    # Add a random feature
    X_with_random = X.copy()
    np.random.seed(random_state)
    X_with_random['random'] = np.random.random(size=len(X))
    
    # Split into train and validation sets
    X_train, X_valid, y_train, y_valid = train_test_split(
        X_with_random, y, test_size=0.2, random_state=random_state, stratify=y
    )
    
    print(f"Train set: {X_train.shape}, Validation set: {X_valid.shape}")
    print(f"Class distribution in train set: {pd.Series(y_train).value_counts().to_dict()}")
    
    # Scale the features
    scaler = StandardScaler()
    X_train_scaled = pd.DataFrame(
        scaler.fit_transform(X_train), columns=X_train.columns, index=X_train.index
    )
    X_valid_scaled = pd.DataFrame(
        scaler.transform(X_valid), columns=X_valid.columns, index=X_valid.index
    )
    
    # Check correlation between features and target
    correlations = {}
    for col in X.columns:
        point_biserial = np.corrcoef(X[col], y)[0, 1]
        correlations[col] = point_biserial
    
    correlations_df = pd.DataFrame({
        'feature': list(correlations.keys()),
        'correlation': list(correlations.values())
    }).sort_values('correlation', key=abs, ascending=False)
    
    print(f"\nTop 10 feature correlations with {target_name}:")
    print(correlations_df.head(10))
    
    # Visualize top correlations
    plt.figure(figsize=(10, 8))
    top_correlations = correlations_df.head(10)
    ax = sns.barplot(x='correlation', y='feature', data=top_correlations, orient='h')
    plt.axvline(x=0, color='black', linestyle='-', linewidth=0.5)
    plt.title(f"Top 10 Feature Correlations with {target_name}", fontsize=16)
    for i, v in enumerate(top_correlations['correlation']):
        ax.text(v + 0.01 if v >= 0 else v - 0.1, i, f'{v:.4f}', va='center')
    plt.tight_layout()
    plt.show()
    
    # Train Random Forest model
    print(f"\nTraining Random Forest classification model for {target_name}...")
    rf = RandomForestClassifier(
        n_estimators=100,
        n_jobs=-1,
        oob_score=True,
        random_state=random_state,
        class_weight='balanced'
    )
    rf.fit(X_train_scaled, y_train)
    
    # Evaluate model performance
    y_train_pred = rf.predict(X_train_scaled)
    y_valid_pred = rf.predict(X_valid_scaled)
    y_train_proba = rf.predict_proba(X_train_scaled)[:, 1]
    y_valid_proba = rf.predict_proba(X_valid_scaled)[:, 1]
    
    train_acc = accuracy_score(y_train, y_train_pred)
    valid_acc = accuracy_score(y_valid, y_valid_pred)
    train_auc = roc_auc_score(y_train, y_train_proba)
    valid_auc = roc_auc_score(y_valid, y_valid_proba)
    oob_score = rf.oob_score_
    
    print(f"Training Accuracy: {train_acc:.4f}")
    print(f"Validation Accuracy: {valid_acc:.4f}")
    print(f"Training AUC-ROC: {train_auc:.4f}")
    print(f"Validation AUC-ROC: {valid_auc:.4f}")
    print(f"OOB Score: {oob_score:.4f}")
    
    # Show confusion matrix and classification report
    print(f"\nConfusion Matrix for {target_name} (Validation Set):")
    cm = confusion_matrix(y_valid, y_valid_pred)
    print(cm)
    
    # Plot confusion matrix
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.title(f'Confusion Matrix for {target_name}')
    plt.show()
    
    # Classification report
    class_names = ['Negative', 'Positive'] if target_name == 'FDA Approval' else ['Non-Toxic', 'Toxic']
    print(f"\nClassification Report for {target_name}:")
    print(classification_report(y_valid, y_valid_pred, target_names=class_names))
    
    # Default Scikit-learn feature importance
    print(f"\nCalculating default Scikit-learn feature importances for {target_name}...")
    base_imp = imp_df(X_train.columns, rf.feature_importances_)
    print(f"Top 10 features ranked by default importance for {target_name}:")
    print(base_imp.head(10))
    
    # Plot features
    var_imp_plot(base_imp.head(15), f'Default feature importance for {target_name} - Top 15')
    
    # Analyze random feature in default importance
    default_random_analysis = analyze_random_feature(base_imp)
    print(f"\nRandom feature in default importance for {target_name}:")
    print(f"Rank: {default_random_analysis['rank']} out of {default_random_analysis['total_features']} features")
    print(f"Top percentile: {default_random_analysis['percentile']:.2f}%")
    print(f"Importance value: {default_random_analysis['importance_value']:.6f}")
    
    # Permutation feature importance
    print(f"\nCalculating permutation feature importance for {target_name}...")
    perm_result = permutation_importance(
        rf, X_train_scaled, y_train, 
        n_repeats=10, random_state=random_state,
        scoring='roc_auc'
    )
    perm_imp_sklearn = imp_df(X_train.columns, perm_result.importances_mean)
    print(f"Top 10 features ranked by permutation importance for {target_name}:")
    print(perm_imp_sklearn.head(10))
    
    # Plot features
    var_imp_plot(perm_imp_sklearn.head(15), f'Permutation feature importance for {target_name} - Top 15')
    
    # Analyze random feature in permutation importance
    perm_random_analysis = analyze_random_feature(perm_imp_sklearn)
    print(f"\nRandom feature in permutation importance for {target_name}:")
    print(f"Rank: {perm_random_analysis['rank']} out of {perm_random_analysis['total_features']} features")
    print(f"Top percentile: {perm_random_analysis['percentile']:.2f}%")
    print(f"Importance value: {perm_random_analysis['importance_value']:.6f}")
    
    # Drop-column feature importance
    print(f"\nCalculating drop-column feature importance for {target_name}...")
    drop_imp = drop_col_feat_imp(rf, X_train_scaled, y_train, metric='roc_auc')
    drop_imp_df = imp_df(X_train.columns, drop_imp)
    print(f"Top 10 features ranked by drop-column importance for {target_name}:")
    print(drop_imp_df.head(10))
    
    # Plot features
    var_imp_plot(drop_imp_df.head(15), f'Drop Column feature importance for {target_name} - Top 15')
    
    # Analyze random feature in drop-column importance
    drop_random_analysis = analyze_random_feature(drop_imp_df)
    print(f"\nRandom feature in drop-column importance for {target_name}:")
    print(f"Rank: {drop_random_analysis['rank']} out of {drop_random_analysis['total_features']} features")
    print(f"Top percentile: {drop_random_analysis['percentile']:.2f}%")
    print(f"Importance value: {drop_random_analysis['importance_value']:.6f}")
    
    # LIME for observation-level feature importance
    print(f"\nCalculating LIME for observation-level feature importance for {target_name}...")
    
    # Find one clear example of each class with high prediction confidence
    train_indices = X_train.index.tolist()
    train_predictions = list(zip(train_indices, y_train, y_train_proba))
    
    # Best positive case (high probability of class 1)
    positive_samples = [(idx, prob) for idx, label, prob in train_predictions if label == 1]
    if positive_samples:
        best_positive_idx, best_positive_prob = max(positive_samples, key=lambda x: x[1])
        
        # Best negative case (low probability of class 1)
        negative_samples = [(idx, prob) for idx, label, prob in train_predictions if label == 0]
        if negative_samples:
            best_negative_idx, best_negative_prob = min(negative_samples, key=lambda x: x[1])
            
            print(f"Selected positive {target_name} example index: {best_positive_idx}, predicted prob: {best_positive_prob:.4f}")
            print(f"Selected negative {target_name} example index: {best_negative_idx}, predicted prob: {best_negative_prob:.4f}")
            
            # Create LIME explainer
            explainer = lime.lime_tabular.LimeTabularExplainer(
                X_train_scaled.values,
                mode='classification',
                feature_names=X_train_scaled.columns,
                class_names=class_names,
                discretize_continuous=True
            )
            
            # Get the scaled values for the selected compounds
            positive_scaled = X_train_scaled.loc[best_positive_idx].values
            negative_scaled = X_train_scaled.loc[best_negative_idx].values
            
            # Explain the positive example
            np.random.seed(random_state)
            exp_positive = explainer.explain_instance(
                positive_scaled, 
                rf.predict_proba, 
                num_features=15  # Top 15 features
            )
            positive_explanation = exp_positive.as_list(label=1)  # Label 1 is the positive class
            print(f"\nLIME explanation for a positive {target_name} example (what makes it positive):")
            for feature, importance in positive_explanation:
                print(f"{feature}: {importance:.4f}")
            
            # Explain the negative example
            np.random.seed(random_state)
            exp_negative = explainer.explain_instance(
                negative_scaled, 
                rf.predict_proba, 
                num_features=15  # Top 15 features
            )
            negative_explanation = exp_negative.as_list(label=1)  # Label 1 is the positive class
            print(f"\nLIME explanation for a negative {target_name} example (what makes it negative):")
            for feature, importance in negative_explanation:
                print(f"{feature}: {importance:.4f}")
            
            # Visualize LIME explanations
            plot_lime_explanation(positive_explanation, f'LIME Explanation for Positive {target_name} Example')
            plot_lime_explanation(negative_explanation, f'LIME Explanation for Negative {target_name} Example')
            
            # Compare the three importance methods side by side (top 10 features for better visualization)
            print(f"\nComparing top 10 features across importance methods for {target_name}:")
            top_features = list(base_imp.head(10)['feature'])
            
            importance_comparison = pd.DataFrame({
                'Feature': top_features,
                'Default': [base_imp[base_imp['feature'] == feat]['feature_importance'].values[0] 
                           for feat in top_features],
                'Permutation': [perm_imp_sklearn[perm_imp_sklearn['feature'] == feat]['feature_importance'].values[0] 
                               for feat in top_features],
                'Drop-Column': [drop_imp_df[drop_imp_df['feature'] == feat]['feature_importance'].values[0] 
                              for feat in top_features]
            })
            
            print(importance_comparison)
            
            # Create a grouped bar chart for method comparison
            plt.figure(figsize=(12, 8))
            x = np.arange(len(top_features))
            width = 0.25
            
            plt.bar(x - width, importance_comparison['Default'], width, label='Default')
            plt.bar(x, importance_comparison['Permutation'], width, label='Permutation')
            plt.bar(x + width, importance_comparison['Drop-Column'], width, label='Drop-Column')
            
            plt.ylabel('Importance Score')
            plt.title(f'Feature Importance Comparison Across Methods - {target_name} - Top 10 Features')
            plt.xticks(x, top_features, rotation=45, ha='right')
            plt.legend()
            plt.tight_layout()
            plt.show()
        else:
            print(f"No negative samples found for {target_name}")
    else:
        print(f"No positive samples found for {target_name}")
    
    print(f"\n{target_name} Analysis Complete")
    print(f"{'='*30}")
    
    return rf, base_imp, perm_imp_sklearn, drop_imp_df


# Main execution code
# Load ClinTox dataset
print("Loading ClinTox dataset...")
df = pd.read_csv('clintox/clintox.csv')
print(f"Total number of compounds: {len(df)}")
print(f"FDA_APPROVED distribution: {df['FDA_APPROVED'].value_counts().to_dict()}")
print(f"CT_TOX distribution: {df['CT_TOX'].value_counts().to_dict()}")

# Calculate molecular descriptors from SMILES
print("\nCalculating molecular descriptors from SMILES...")
descriptors_df, valid_indices = calculate_mol_descriptors(df['smiles'])

if len(valid_indices) == 0:
    raise ValueError("No valid molecules could be processed. Check RDKit installation and SMILES format.")

# Map descriptors to correct targets
valid_df = df.iloc[valid_indices]
valid_fda = valid_df['FDA_APPROVED'].values
valid_ct_tox = valid_df['CT_TOX'].values

print(f"Successfully calculated descriptors for {len(descriptors_df)} compounds out of {len(df)}")
print(f"Valid FDA_APPROVED distribution: {pd.Series(valid_fda).value_counts().to_dict()}")
print(f"Valid CT_TOX distribution: {pd.Series(valid_ct_tox).value_counts().to_dict()}")

# Check for missing or infinite values
X = descriptors_df.copy()
bad_values = X.isna().sum().sum() + np.isinf(X.values).sum()
if bad_values > 0:
    print(f"Found {bad_values} missing or infinite values. Filling with median values.")
    X = X.replace([np.inf, -np.inf], np.nan)
    
    # Fill NaN values with median for each column
    for col in X.columns:
        if X[col].isna().sum() > 0:
            median_val = X[col].median()
            # If median is also NaN, use 0 instead
            if pd.isna(median_val):
                X[col] = X[col].fillna(0)
            else:
                X[col] = X[col].fillna(median_val)

# Run analysis for FDA Approval
print("\nAnalyzing FDA Approval...")
rf_fda, base_imp_fda, perm_imp_fda, drop_imp_fda = analyze_feature_importance(
    X, valid_fda, "FDA Approval"
)

# Run analysis for Clinical Trial Toxicity
print("\nAnalyzing Clinical Trial Toxicity...")
rf_tox, base_imp_tox, perm_imp_tox, drop_imp_tox = analyze_feature_importance(
    X, valid_ct_tox, "Clinical Trial Toxicity"
)

# Compare feature importance between FDA Approval and Toxicity
print("\nComparing top features across both tasks...")

# Get top 10 features from each task
fda_top_features = set(base_imp_fda.head(10)['feature'])
tox_top_features = set(base_imp_tox.head(10)['feature'])

# Find common and unique features
common_features = fda_top_features.intersection(tox_top_features)
fda_unique = fda_top_features - common_features
tox_unique = tox_top_features - common_features

print(f"Common top features between FDA Approval and Toxicity: {common_features}")
print(f"Features unique to FDA Approval: {fda_unique}")
print(f"Features unique to Toxicity: {tox_unique}")

# Check if random feature is important in either task
print("\nRandom feature analysis:")
for task, imp_df in [("FDA Approval", base_imp_fda), ("Clinical Trial Toxicity", base_imp_tox)]:
    random_analysis = analyze_random_feature(imp_df)
    print(f"{task}: Random feature ranks {random_analysis['rank']} out of {random_analysis['total_features']}")

# Calculate total execution time
end_time = time.time()
total_time = end_time - start_time
print(f"\nTotal execution time: {total_time:.2f} seconds ({total_time/60:.2f} minutes)")

print("\nAnalysis complete!")
```
